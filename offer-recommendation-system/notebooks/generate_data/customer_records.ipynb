```python
def generate_customer_ids(n):
 """Generate n unique 10-character alphanumeric IDs"""
 chars = string.ascii_uppercase + string.digits
 ids = set()
 while len(ids) < n:
 new_id = ''.join(random.choices(chars, k=10))
 ids.add(new_id)
 return list(ids)

def generate_postal_code(province):
 """Generate valid Canadian postal code for province"""
 prefix = random.choice(postal_prefixes[province])
 letter1 = random.choice(string.ascii_uppercase.replace('D', '').replace('F', '').replace('I', '').replace('O', '').replace('Q', '').replace('U', ''))
 digit1 = random.choice(string.digits)
 letter2 = random.choice(string.ascii_uppercase.replace('D', '').replace('F', '').replace('I', '').replace('O', '').replace('Q', '').replace('U', ''))
 digit2 = random.choice(string.digits)
 letter3 = random.choice(string.ascii_uppercase.replace('D', '').replace('F', '').replace('I', '').replace('O', '').replace('Q', '').replace('U', ''))
 return f"{prefix}{letter1} {digit1}{letter2}{digit2}{letter3}"

def generate_dob(min_age=18, max_age=85):
 """Generate date of birth with realistic age distribution"""
 # Weight towards working age (25-55)
 if random.random() < 0.6:
 age = random.randint(25, 55)
 elif random.random() < 0.8:
 age = random.randint(18, 24)
 else:
 age = random.randint(56, 85)

 base_date = datetime(2024, 1, 1)
 birth_year = base_date.year - age
 birth_month = random.randint(1, 12)
 birth_day = random.randint(1, 28) # Safe for all months
 return datetime(birth_year, birth_month, birth_day).strftime('%Y-%m-%d')

def generate_phone(province):
 """Generate phone with province-appropriate area code"""
 area_codes = {
 'AB': ['403', '587', '780', '825'],
 'BC': ['236', '250', '604', '672', '778'],
 'SK': ['306', '639'],
 'MB': ['204', '431']
 }
 area = random.choice(area_codes[province])
 return f"{area}-{random.randint(100,999)}-{random.randint(1000,9999)}"
```

```python
def generate_email(first, last):
    """Generate email address"""
    domain = random.choice(email_domains)
    fmt = random.randint(0, 3)
    num = random.randint(1, 999)

    if fmt == 0:
        return f"{first.lower()}.{last.lower()}@{domain}"
    elif fmt == 1:
        return f"{first.lower()}{num}@{domain}"
    elif fmt == 2:
        return f"{first[0].lower()}{last.lower()}@{domain}"
    else:
        return f"{first.lower()}_{last.lower()}{num}@{domain}"

customer_ids = generate_customer_ids(n_unique_customers)

provinces = list(cities_by_province.keys())
province_weights = [0.35, 0.35, 0.15, 0.15]

customers = []
for i in range(n_unique_customers):
    province = random.choices(provinces, weights=province_weights)[0]
    city = random.choice(cities_by_province[province])
    first = random.choice(first_names)
    last = random.choice(last_names)

    customers.append({
        'record_id': f"REC{str(i).zfill(7)}",
        'true_customer_id': customer_ids[i],
        'first_name': first,
        'last_name': last,
        'email': generate_email(first, last),
        'phone': generate_phone(province),
        'date_of_birth': generate_dob(),
        'city': city,
        'province': province,
        'postal_code': generate_postal_code(province),
        'registration_date': (datetime(2020, 1, 1) + timedelta(days=random.randint(0, 1800))).strftime('%Y-%m-%d')
    })

df_unique = pd.DataFrame(customers)
del customers
```

```python
dup_types = np.random.choice(
    ['true_dup', 'name_collision', 'shared_phone', 'shared_email', 'id_error'],
    size=n_duplicates,
    p=[0.40, 0.20, 0.15, 0.15, 0.10]
)

duplicates = []
used_indices = set()

for i, dup_type in enumerate(dup_types):
    source_idx = random.randint(0, n_unique_customers - 1)
    source = df_unique.iloc[source_idx].to_dict()

    dup = source.copy()
    dup['record_id'] = f"REC{str(n_unique_customers + i).zfill(7)}"

    if dup_type == 'true_dup':
        variation = random.choice(['typo_first', 'typo_last', 'diff_email', 'missing_phone', 'typo_city'])

        if variation == 'typo_first' and len(dup['first_name']) > 2:
            name = dup['first_name']
            typo_type = random.choice(['drop_letter', 'swap', 'double'])
            if typo_type == 'drop_letter':
                pos = random.randint(1, len(name)-1)
                dup['first_name'] = name[:pos] + name[pos+1:]
            elif typo_type == 'swap' and len(name) > 2:
                pos = random.randint(0, len(name)-2)
                dup['first_name'] = name[:pos] + name[pos+1] + name[pos] + name[pos+2:]
            else:
                pos = random.randint(0, len(name)-1)
                dup['first_name'] = name[:pos] + name[pos] + name[pos:]

        elif variation == 'typo_last' and len(dup['last_name']) > 2:
            dup['last_name'] = dup['last_name'][:-1]

        elif variation == 'diff_email':
            dup['email'] = generate_email(dup['first_name'], dup['last_name'])

        elif variation == 'missing_phone':
            dup['phone'] = None

        elif variation == 'typo_city':
            city = dup['city']
            if len(city) > 3:
                dup['city'] = city[:-1] + random.choice('aeiou')

    elif dup_type == 'name_collision':
        common_first = random.choice(['John', 'James', 'Michael', 'David', 'Robert', 'Mary', 'Jennifer', 'Sarah', 'Lisa', 'Jessica'])
        common_last = random.choice(['Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Davis', 'Miller', 'Wilson', 'Moore', 'Taylor'])

        dup['true_customer_id'] = ''.join(random.choices(string.ascii_uppercase + string.digits, k=10))
        dup['first_name'] = common_first
        dup['last_name'] = common_last
        province = random.choices(provinces, weights=province_weights)[0]
        dup['email'] = generate_email(common_first, common_last)
        dup['phone'] = generate_phone(province)
        dup['date_of_birth'] = generate_dob()
        dup['city'] = random.choice(cities_by_province[province])
        dup['province'] = province
        dup['postal_code'] = generate_postal_code(province)

    elif dup_type == 'shared_phone':
        dup['true_customer_id'] = ''.join(random.choices(string.ascii_uppercase + string.digits, k=10))
        dup['first_name'] = random.choice(first_names)
        dup['last_name'] = source['last_name'] if random.random() < 0.5 else random.choice(last_names)
        dup['email'] = generate_email(dup['first_name'], dup['last_name'])
        dup['date_of_birth'] = generate_dob()

    elif dup_type == 'shared_email':
        dup['true_customer_id'] = ''.join(random.choices(string.ascii_uppercase + string.digits, k=10))
        dup['first_name'] = random.choice(first_names)
        dup['last_name'] = source['last_name'] if random.random() < 0.5 else random.choice(last_names)
        dup['phone'] = generate_phone(dup['province'])
        dup['date_of_birth'] = generate_dob()

    elif dup_type == 'id_error':
        dup['first_name'] = random.choice(first_names)
        dup['last_name'] = random.choice(last_names)
        province = random.choices(provinces, weights=province_weights)[0]
        dup['email'] = generate_email(dup['first_name'], dup['last_name'])
        dup['phone'] = generate_phone(province)
        dup['date_of_birth'] = generate_dob()
        dup['city'] = random.choice(cities_by_province[province])
        dup['province'] = province
        dup['postal_code'] = generate_postal_code(province)

    duplicates.append(dup)

df_duplicates = pd.DataFrame(duplicates)
del duplicates
```

```python
df_customers = pd.concat([df_unique, df_duplicates], ignore_index=True)
del df_unique, df_duplicates

df_customers = df_customers.sample(frac=1, random_state=42).reset_index(drop=True)
df_customers['record_id'] = [f"REC{str(i).zfill(7)}" for i in range(len(df_customers))]

print(f"Total: {len(df_customers):,} records | Unique: {df_customers['true_customer_id'].nunique():,}")

df_customers['province'].value_counts()
```
